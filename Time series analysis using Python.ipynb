```python
cols = ['Row ID', 'Order ID', 'Ship Date', 'Ship Mode', 'Customer ID', 'Customer Name', 'Segment', 'Country', 'City', 'State',
       'Postal Code', 'Region', 'Product ID', 'Category', 'Sub-Category', 'Product Name', 'Quantity', 'Discount', 'Profit']
furniture.drop(cols, axis= 1, inplace=True)
furniture = furniture.sort_values('Order Date')
furniture.isnull().sum()#Checking for null values
```

```python
furniture.head()
```

```python
furniture = furniture.groupby("Order Date")['Sales'].sum().reset_index()#sumof all sales acc.. to date (Order date)
furniture
```

```python
furniture = furniture.set_index('Order Date')
furniture.index
```

```python
y = furniture['Sales'].resample('MS').mean()
y
```

```python
#visualizing furniture sales time series data

y.plot(figsize =(15, 6))
plt.show()
```

```python
#ARIMA TIME SERIES ANALYSIS

p=d=q = range(0, 2)# here p, d, q are the integers
# Here auto regressive part(p), integrated part(d),moving average(q)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
print("SARIMAX:{} X {}".format(pdq[1], seasonal_pdq[1]))
print("SARIMAX:{} X {}".format(pdq[2], seasonal_pdq[2]))
print("SARIMAX:{} X {}".format(pdq[3], seasonal_pdq[3]))
print("SARIMAX:{} X {}".format(pdq[4], seasonal_pdq[4]))

#These are the different patterns from 0 to 12
```

```python
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(y, order = param, seasonal_order = param_seasonal, enforce_stationarity = False, 
                                           enforce_invertibility= False)
            result = mod.fit()
            print("ARIMA{} X{} 12 -AIC: {}".format(param, param_seasonal, result.aic))
        except:
            continue
```

```python
mod = sm.tsa.statespace.SARIMAX(y, order = (1, 1, 1), seasonal_order = (1, 1, 0, 12), 
                               enforce_stationarity = False, enforce_invertibility = False)
results = mod.fit()
print(results.summary().tables[1])
```

```python
# FORECASTING 
pred = results.get_prediction(start = pd. to_datetime('2017-01-01'), dynamic = False)
pred_future = pred.conf_int()
ax = y['2014':].plot(label = 'Observed')
pred.predicted_mean.plot(ax=ax, label = 'One step ahead Forecast', alpha = 0.7, figsize= (14, 7))
ax.fill_between(pred_future.index, pred_future.iloc[:, 0],
               pred_future.iloc[:, 1], color = 'k', alpha = 0.2)
ax.set_xlabel("Date")
ax.set_ylabel("Furniture Sales")
plt.legend()
plt.show()


# Here is the predicted data of future from Jan 2017 to ....... in the form of one step ahead forecast
```

```python
y_forecasted = pred.predicted_mean
y_truth = y["2017-01-01":]
#compute the mean square error(MSE)
mse = ((y_forecasted - y_truth)** 2).mean()
print("Mean squared error of forecast is {}".format(round(mse, 2)))
print("Root Mean squared error of forecast is {}".format(round(np.sqrt(mse), 2)))
```

```python
# FORECASTING 
pred_f = results.get_forecast(steps = 100)
pred_future = pred_f.conf_int()
ax = y.plot(label = 'Observed', figsize= (14, 7))
pred_f.predicted_mean.plot(ax=ax, label = 'Forecast')
ax.fill_between(pred_future.index, pred_future.iloc[:, 0],
               pred_future.iloc[:, 1], color = 'k', alpha = 0.2)
ax.set_xlabel("Date")
ax.set_ylabel("Furniture Sales")
plt.legend()
plt.show()


# Here is the predicted data of future from Jan 2017 to ....... in the form of one step ahead forecast
```
